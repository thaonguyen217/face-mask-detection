```
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('1626.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_h, img_w = img.shape[:2]
bboxs = []

with open('1626.txt', 'r') as f:
  bbxs = f.readlines()
for bbx in bbxs:
  bbox = bbx.strip().split()
  label = int(bbox[0])
  x_center = float(bbox[1])
  y_center = float(bbox[2])
  w = float(bbox[3])
  h = float(bbox[4])
  xmin = int((x_center - w/2)*img_w)
  ymin = int((y_center - h/2)*img_h)
  xmax = xmin + int(w*img_w)
  ymax = ymin + int(h*img_h)
  cv2.rectangle(img, (xmin, ymin), (xmax, ymax), [0,255,0], 2)
  cv2.putText(img, str(label), (xmin, ymin-3), cv2.FONT_HERSHEY_COMPLEX, 0.6, [0,255,0], 2)
plt.imshow(img)
plt.axis('off')
plt.show()
```
